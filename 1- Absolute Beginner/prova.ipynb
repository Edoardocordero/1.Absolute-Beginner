```python
import pandas as pd
from pandas_datareader import data as web
import numpy as np
import matplotlib.pyplot as plt
```

```python
tlist=['AMZN', 'MSFT', 'GOOGL', 'AAPL']
pesi=[0.4, 0.2, 0.3, 0.1]
```

```python
data=pd.DataFrame()
for t in tlist:
    data[t]= web.DataReader(t, 'yahoo', '2018-9-1')['Adj Close']
```

```python
data_norm= data/data.iloc[0] *100
```

```python
port_norm=np.dot(data_norm, pesi)
```

```python
data_ret= np.log(data/data.shift(1)) 
```

```python
port_ret=np.dot(data_ret, pesi)
```

```python
data_norm.plot(figsize= (16, 5))
plt.show()
```

```python
port_norm= pd.DataFrame(port_norm)
```

```python
port_norm.plot( figsize= (16, 5) )
plt.show()
```

```python
port_ret[0]= 0
```

```python
print('Rendimento medio annuo: ',round( port_ret.mean()*250*100,3), '%')
```
